```python
from google.colab import files

# This cell handles the uploading of a FASTA file from the user's local machine.
# It prompts the user to upload a file, reads its content, and stores it for further processing.
print("Please upload your FASTA file:")
uploaded_fasta = files.upload()

# Get the name of the uploaded file
fasta_file_name = next(iter(uploaded_fasta))

# Read the content of the FASTA file
fasta_content = uploaded_fasta[fasta_file_name].decode('utf-8')

print(f"Successfully uploaded '{fasta_file_name}'.")
```

```python
# Parse the FASTA content
def parse_fasta(fasta_string):
    """
    Parses a FASTA format string into a list of dictionaries.

    Each dictionary contains 'header' and 'sequence' keys.

    Args:
        fasta_string (str): A string containing FASTA formatted sequences.

    Returns:
        list: A list of dictionaries, where each dictionary represents a sequence
              with its header and sequence content.
    """
    sequences = []
    header = ''
    current_sequence = ''
    for line in fasta_string.splitlines():
        line = line.strip()
        if not line:
            continue
        if line.startswith('>'):
            if header and current_sequence:
                sequences.append({'header': header, 'sequence': current_sequence})
            header = line[1:].strip()
            current_sequence = ''
        else:
            current_sequence += line
    if header and current_sequence:
        sequences.append({'header': header, 'sequence': current_sequence})
    return sequences

parsed_fasta_data = parse_fasta(fasta_content)

print(f"Parsed {len(parsed_fasta_data)} sequences from '{fasta_file_name}'.")
```

```python
# This cell displays the first few parsed sequences to give a preview of the data.
# It shows the header and the beginning of the sequence for up to 5 entries.
if parsed_fasta_data:
    print("\nFirst 5 entries:")
    for i, entry in enumerate(parsed_fasta_data[:5]):
        print(f"Entry {i+1}:")
        print(f"  Header: {entry['header']}")
        print(f"  Sequence (first 50 chars): {entry['sequence'][:50]}{'...' if len(entry['sequence']) > 50 else ''}")
else:
    print("No sequences found in the FASTA file.")
```

```python
def calculate_gc(sequence_header, sequence_string):
    """
    Calculates the GC content percentage of a DNA/RNA sequence.

    Args:
        sequence_header (str): The header of the sequence (for identification).
        sequence_string (str): The nucleotide sequence string.

    Returns:
        tuple: A tuple containing the sequence header (str) and its GC content (float).
               Returns 0.0 if the sequence is empty or contains no valid bases.
    """
    if not sequence_string:
        return sequence_header, 0.0

    # Convert to uppercase to handle mixed-case sequences
    sequence_string = sequence_string.upper()

    gc_count = sequence_string.count('G') + sequence_string.count('C')
    total_bases = len(sequence_string)

    if total_bases == 0:
        return sequence_header, 0.0

    gc_content = (gc_count / total_bases) * 100
    return sequence_header, gc_content

# Calculate GC content for all parsed sequences
# This loop iterates through the parsed FASTA data and applies the calculate_gc function
# to each sequence, then prints the header and calculated GC content.
print("GC Content for each sequence:")
for entry in parsed_fasta_data:
    header, gc_percentage = calculate_gc(entry['header'], entry['sequence'])
    print(f"Header: {header}\nGC Content: {gc_percentage:.2f}%\n")
```
